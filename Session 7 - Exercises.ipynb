## Problem 2
Write a program to detect a Palindrome name that when you read it from left or right will be the same.

Example:

 Aziza-True

 Mawada-False

```python
txt2 = "Aziza"
l=txt2.lower()
if l[::-1]==l[0::]:
    print("palindrome name")
else:
    print("not palindrome name")    

```
